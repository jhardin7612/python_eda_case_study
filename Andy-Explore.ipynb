## Imports and creating dataframe

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
nutri_df = pd.read_csv('data/health_nutrition.csv')
nutri_df.head()
```

## In case we wanted to use GNI I filtered out just GNI here

```python
nutri_df.loc[(nutri_df['Indicator Name']=='GNI per capita, Atlas method (current US$)')]
```

Here I wanted to remove all the columns I did not need, so its a loop creating a list from 1960 to 2014 and passing that list to be dropped from the dataframe

```python
nlst = []
for i in range(1960,2015):
    nlst.append(str(i))
nlst.append('Unnamed: 60')
just_15_nutri_df = nutri_df.drop(columns=nlst)
```

```python
nlst = []
for i in range(1960,2010):
    nlst.append(str(i))
nlst.append('Unnamed: 60')
five_year_nutri_df = nutri_df.drop(columns=nlst)
```

Just want the Baltic countries for the baltic_df so filtered it by country code.
Then replaced any NaN value with 0 and filtered out all 0s.

```python
baltic_df = just_15_nutri_df[just_15_nutri_df['Country Code']=='CEB']
drop_na_baltic_df = baltic_df.fillna(0)
drop_na_baltic_df = drop_na_baltic_df[drop_na_baltic_df['2015']!= 0]
```

```python
drop_na_baltic_df.set_index('Indicator Name')
baltic_filtered_df = drop_na_baltic_df[drop_na_baltic_df['Indicator Name'].str.contains('GNI')]
baltic_filtered_df
```

## Want to filter out Balkan countries to make seperate dataframe of just Balkan Data

```python
balkans_df = five_year_nutri_df[(just_15_nutri_df['Country Code']=='ALB') |
                              (just_15_nutri_df['Country Code']=='BIH') |
                              (just_15_nutri_df['Country Code']=='BGR') |
                              (just_15_nutri_df['Country Code']=='GRC') |
                              (just_15_nutri_df['Country Code']=='KSV') |
                              (just_15_nutri_df['Country Code']=='MNE') |
                              (just_15_nutri_df['Country Code']=='MKD') |
                              (just_15_nutri_df['Country Code']=='TUR') |
                              (just_15_nutri_df['Country Code']=='SRB') |
                              (just_15_nutri_df['Country Code']=='HRV') ]
balkans_df
```

# Here I want to drop all NaNs just like the Baltic countries.

```python
drop_na_balkans_df = balkans_df.fillna(0)
drop_na_balkans_df = drop_na_balkans_df[drop_na_balkans_df['2015']!= 0]
```

```python
balkans_filtered_df = drop_na_balkans_df[drop_na_balkans_df['Indicator Name'].str.contains('GNI')]
balkans_filtered_df.plot(x='Country Name',y=['2010','2011','2012','2013','2014','2015'],kind='bar', title='Balkan GNI by country')
balkans_filtered_df
```

# Since Balkan have all the countries listed out while Baltic does not, I will take the mean GNI for 

```python

balkans_filtered_df['means'] = balkans_filtered_df[['2010','2011','2012','2013','2014','2015']].mean(numeric_only=True, axis=1)
balkans_filtered_df
```

```python
sns.set_theme(rc={'figure.figsize':(15,15)})
sns.barplot(x=balkans_filtered_df.index,y=balkans_filtered_df['means']).set_title('Means of GNI Over Last 5 Years in Dataset')
```

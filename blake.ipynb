```python

df_filter = (df['Country Name'] == 'Lithuania') | (df['Country Name'] == 'Latvia') | (df['Country Name'] == 'Estonia') | (df['Country Name'] == 'Bosnia') | (df['Country Name'] == 'Serbia') | (df['Country Name'] == 'Bulgaria') | (df['Country Name'] == 'Albania') | (df['Country Name'] == 'Montenegro') | (df['Country Name'] == 'North Macedonia') | (df['Country Name'] == 'Kosovo') | (df['Country Name'] == 'Greece') | (df['Country Name'] == 'Croatia') | (df['Country Name'] == 'Romania') | (df['Country Name'] == 'Turkey') | (df['Country Name'] == 'Slovenia')
df2  = df[df_filter]
df3 = df2.T
df3.head()
```

```python
df4 = df3.drop(['Indicator Code','Country Code'])
df4.columns = df3.loc['Indicator Name']
#bool_mask = df3.loc['Country Name'] == [['Lithuania', 'Latvia', 'Estonia', 'Bosnia', 'Serbia', 'Bulgaria', 'Albania', 'Montenegro', 'North Macedonia','Kosovo', 'Greece','Croatia', 'Romania', 'Turkey', 'Slovenia']]
```

```python
df4 = df4[['Health expenditure, total (% of GDP)','Health expenditure per capita (current US$)','GNI per capita, Atlas method (current US$)','Adults (ages 15+) and children (0-14 years) living with HIV']]
df4 = df4.drop('Country Name')
df4 = df4.drop('Indicator Name')
df4 = df4.dropna(axis=1, how='all')
df4.head()
```

```python
df4['Health expenditure, total (% of GDP)'] = df4[df4.columns[0:12]].mean(axis=1)
df4['Health expenditure per capita (current US$)'] = df4[df4.columns[12:24]].mean(axis=1)
df4['GNI per capita, Atlas method (currnet US$)'] = df4[df4.columns[24:37]].mean(axis=1)
df4['Adults (ages 15+) and children (0-14 years) living with HIV'] = df4[df4.columns[37:]].mean(axis=1)
final_df = df4[['Health expenditure, total (% of GDP)','Health expenditure per capita (current US$)','GNI per capita, Atlas method (currnet US$)','Adults (ages 15+) and children (0-14 years) living with HIV']]
final_df = final_df.drop('Unnamed: 60')
final_df
```
